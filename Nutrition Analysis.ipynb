```python
hypo1 = nut_df["T3 ng/Ml"]
bins = [1.0,1.2,1.4,1.6,1.8]
labels = ['1.01-1.20', '1.21-1.40', '1.41-1.60', '1.61-1.81']
nut_df['T3_range'] = pd.cut(nut_df['T3 ng/Ml'], bins= bins, labels= labels, right= False)
```

# T4

    T4 = 3.2 - 12.6 Âµg/dl


```python
hypo4 = nut_df["T4 ug/Ml"]
hypo4.unique()
```

```python
bins = [3.0, 5.0, 7.0, 9.0, 11.0, 13.0]
labels = ['3.01-5.00', '5.01-7.00', '7.01-9.00', '9.01-11.00', '11.01-13.00']
nut_df['T4_range'] = pd.cut(nut_df["T4 ug/Ml"], bins= bins, labels= labels, right= False)
```

```python
t4_rng = round((nut_df['T4_range'].value_counts(normalize=True)*100),2)
t4_rng
```

```python
plt.figure(figsize=(15,5))
plt.title("T4 Hypothyroidism History", fontsize= 15, fontweight= 'bold')
clrs = ('red' if (i == t4_rng.index[4]) else 'seagreen' for i in t4_rng.index)
t4 = sns.barplot(y= t4_rng.index, x= t4_rng.values,orient= 'h', palette= clrs)
plt.ylabel("T4 Range", fontsize= 10, fontweight= 'bold')
plt.xlabel("Frequency", fontsize= 10, fontweight= 'bold')
t4.bar_label(t4.containers[0], fmt = '%.1f%%', fontweight= 'bold', color= '#660033')
sns.despine(left=True, bottom=True)
plt.show()
```
